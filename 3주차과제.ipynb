#### 4-2. matplotlib 라이브러리 자유자재로 사용하기
기초 그래프 그리기
1. tips 데이터 불러오기

```python
tips = sns.load_dataset("tips")
print(tips.head())
print(type(tips))
```

2. 히스토그램 그리기

```python
fig = plt.figure()  # 기본 틀 마련
axes1 = fig.add_subplot(1, 1, 1)  # 그래프 격자(axes1) 구성성
axes1.hist(tips['total_bill'], bins=10)  # total_bill이 이 그래프의 변수, x축 간격 10으로 지정정
axes1.set_title('Histogram of Total Bill')
axes1.set_xlabel('Frequency')
axes1.set_ylabel('Total Bill')
```

3. 산점도 그래프 기리기 (이변량 그래프)

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)
axes1.scatter(tips['total_bill'], tips['tip'])
axes1.set_title('Scatterplot of Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

4. 박스 그래프 그리기 (이산형 변수와 연속형 변수 함께 사용

```python
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1, 1, 1)
axes1.boxplot([tips[tips['sex']=='Female']['tip'],
               tips[tips['sex']=='Male']['tip']],
              labels=['Female', 'Male'])
axes1.set_title('Boxplot of Tips by Sex')
axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
```

다변량 그래프 그리기 (3개 이상의 변수 사용한 그래프)
1. 여성 0, 남성 1로 반환

```python
# 문자열 정수로 치환하는 함수 
def recode_sex(sex):
  if sex == 'Female':
    return 0
  else:
    return 1
```

2. 반환값 데이터프레임에 추가

```python
tips['sex_color'] = tips['sex'].apply(recode_sex)
```

3. 테이블당 인원 수(size) 데이터 추가 후 scatter plot 그리기

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)
axes1.scatter(
    x = tips['total_bill'],
    y = tips['tip'],
    s = tips['size']*10,  # 점 크기
    c = tips['sex_color'],
    alpha = 0.5)  # 투명도
axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

#### 4-3. seaborn 라이브러리 자유자재로 사용하기
1. 데이터 및 라이브러리 불러오기

```python
import seaborn as sns
tips = sns.load_dataset("tips")
```

2. 히스토그램 그리기

```python
ax = plt.subplot()  # 기본 틀
ax = sns.distplot(tips['total_bill'])  # 밀집도 그래프도 함께 그려줌
ax.set_title('Total Bill Histogram with Density plot')
```

3. 밀집도 그래프 제외: kde=False

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], kde=False) 
ax.set_title('Total Bill Histogram ')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

4. 밀집도 그래프만 나타내기: hist=False

```python
ax = plt.subplots() 
ax = sns.distplot(tips['total_bill'], hist=False) 
ax.set_title('Total Bill Histogram ')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Unit Probability')
```

5. 양탄자(rug) 그래프 그리기
* 양탄자 그래프: 그래프의 축에 동일한 길이의 직선을 붙여 데이터의 밀집 정도를 표현한 그래프

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], rug=True)
ax.set_title('Total Bill Histogram with Density and Rug Plot')
ax.set_xlabel('Total Bill')
```

6. count 그래프 그리기 (이산값)

```python
ax = plt.subplots()
ax = sns.countplot('day', data=tips)
ax.set_title('Count of days')
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Frequency')
```

다양한 종류의 이변량 그래프 그리기

1. seaborn 라이브러리로 산점도 그래프 그리기

  regplot 메서드 사용: 산점도 그래프와 회귀선 함께 그릴 수 있음

```python
ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
# 회귀선 제거: fit_reg 인자를 False로 지정
ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=False)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

2. 산점도 그래프와 히스토그램 한 번에 그리기

  jointplot 메서드 사용

```python
joint = sns.jointplot(x='total_bill', y='tip', data=tips)
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

3. 육각 그래프 그리기

  jointplot 메서드에서 kind 인잣값을 hex로 지정

```python
hexbin = sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
hexbin.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

4. 이차원 밀집도 그리기

  kdeplot 메서드

```python
ax = plt.subplots()
ax = sns.kdeplot(data=tips['total_bill'],
                 data2=tips['tip'],
                 shade=True)
ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

5. 바 그래프 그리기

```python
ax = plt.subplots()
ax = sns.barplot(x='time', y='total_bill', data=tips)
ax.set_title('Bar plot of average total bill for time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total bill')
```

6. 박스 그래프 그리기

```python
ax = plt.subplots()
ax = sns.boxplot(x='time', y='total_bill', data=tips)
ax.set_title('Boxplot of total bill by time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Total Bill')
```

7. 바이올린 그래프 그리기

  박스 그래프는 데이터 분산이 모호하게 표현됨 이런 경우 바이올린 그래프 사용(박스 형태를 바이올린 형태로)


```python
ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', data=tips)
ax.set_title('Violin plot of total bill for time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Total Bill')
```

8. 관계 그래프 그리기

  지금까지의 그래프 종합한 그래프

  pairplot 메서드에 데이터프레임 넣는 방법

```python
fig = sns.pairplot(tips)
# 대각선을 기준으로 위 아래가 같음
```

```python
pair_grid = sns.PairGrid(tips)
pair_grid = pair_grid.map_upper(sns.regplot)  # map_upper 메서드는 대각선을 기준으로 위쪽에 그릴 그래프 지정, 산점도도
pair_grid = pair_grid.map_lower(sns.kdeplot)  # map_lower 메서드는 대각선을 기준으로 아래쪽에 그릴 그래프 지정, 이차원밀집도
pair_grid = pair_grid.map_diag(sns.distplot, rug=True)  # 히스토그램, 밀집도
plt.show()
```

#### 4-5. seaborn 라이브러리로 그래프 스타일 설정하기

```python
fig, ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True)
```

```python
# whitegrid로 스타일 설정: 가로줄
sns.set_style('whitegrid')
fig, ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True)
```

```python
# for문 이용하여 모든 스타일 하나씩 적용
fig = plt.figure()

seaborn_styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']

for idx, style in enumerate(seaborn_styles):
  plot_position = idx+1
  with sns.axes_style(style):
    ax = fig.add_subplot(2, 3, plot_position)
    violin = sns.violinplot(x='time', y='total_bill', data=tips, ax=ax)
    violin.set_title(style)

fig.tight_layout()
```
